```python
def bar_diagram(x, y, colores):
    fig, ax = plt.subplots()
    ax.bar(x , y, color = colores)
    plt.show()
    
subjects = {'Mates': 7, 'Fsica': 3, 'Lengua': 9}
x = subjects.keys()
y = subjects.values()
colores = ["#619cff", "#00ba38", "#f8766d"]
bar_diagram(x, y, colores)
```
