```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Crear un diccionario con los datos
data = {
'Fecha': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
'Producto': ['A', 'B', 'A', 'C', 'B'],
'Cantidad': [10, 15, 8, 12, 20],
'Precio unitario': [5.99, 7.5, 6.25, 9.99, 8.75],
'Total': [59.9, 112.5, 50, 119.88, 175]
}
# Crear el DataFrame a partir del diccionario
df = pd.DataFrame(data)
# Mostrar el DataFrame
print(df)
```

```python
# EJERCICIO 2
```

```python
print(df.head())
print('\n')
print(df.info())
print('\n')
print(df.describe())
```

```python
# EJERCICIO 3
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
# EJERCICIO 4
```

```python
# Cantidad
plt.hist(df['Cantidad'], bins=10, color='b')
plt.title('Histograma de Cantidad')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.show()

# Precio unitario
plt.hist(df['Precio unitario'], bins=10, color='r')
plt.title('Histograma de Precio unitario')
plt.xlabel('Precio unitario')
plt.ylabel('Frecuencia')
plt.show()

# Total
plt.hist(df['Total'], bins=10, color='g')
plt.title('Histograma de Total')
plt.xlabel('Total')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# EJERCICIO 5
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Producto', y='Cantidad', hue='Producto', data=df, palette='Set1')
plt.title('Boxplot de Cantidad por Producto (con separación por color)')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```

```python
# EJERCICIO 6
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
